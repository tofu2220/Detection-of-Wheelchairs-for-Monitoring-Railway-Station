```python
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
image_path = 'dataset/images/0a010cf198105d4f.jpg'
image = Image.open(image_path)
```

```python
image_width, image_height = image.size
```

```python
label_path = 'dataset/labels/0a010cf198105d4f.txt'
with open(label_path, 'r') as f:
    label_data = f.readline().strip().split()
```

```python
class_name = label_data[0]
x_center = float(label_data[1])
y_center = float(label_data[2])
width = float(label_data[3])
height = float(label_data[4])
```

```python
x1 = int((x_center - width / 2) * image_width)
y1 = int((y_center - height / 2) * image_height)
x2 = int((x_center + width / 2) * image_width)
y2 = int((y_center + height / 2) * image_height)
```

```python
plt.imshow(image)
plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2, edgecolor='r', facecolor='none'))
plt.text(x1, y1, class_name, bbox=dict(facecolor='red', alpha=0.5))
plt.axis('off')
plt.show()
```

```python
dataset_dir = 'dataset'

image_dir = os.path.join(dataset_dir, 'images')
image_files = os.listdir(image_dir)[6:9]

for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    image = Image.open(image_path)

    image_width, image_height = image.size

    label_path = os.path.join(dataset_dir, 'labels', image_file[:-3] + 'txt')
    with open(label_path, 'r') as f:
        label_data = f.readline().strip().split()

    class_name = label_data[0]
    x_center = float(label_data[1])
    y_center = float(label_data[2])
    width = float(label_data[3])
    height = float(label_data[4])

    x1 = int((x_center - width / 2) * image_width)
    y1 = int((y_center - height / 2) * image_height)
    x2 = int((x_center + width / 2) * image_width)
    y2 = int((y_center + height / 2) * image_height)

    plt.imshow(image)
    plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2, edgecolor='r', facecolor='none'))
    plt.text(x1, y1, class_name, bbox=dict(facecolor='red', alpha=0.5))
    plt.axis('off')
    plt.title(image_file)
    plt.show()
```
